```python
# Librerias y Bibliotecas que necesitaremos
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
import sys
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Se guardan los nombres de los archivos en una lista
dirs = glob.glob('estudiantes.csv')
file = pd.DataFrame()

#Se crean tablas de pandas para concatenar la informaci√≥n
for i in dirs:
    file = pd.read_csv(i,sep = ',',thousands='.',skiprows=1, header=None, na_values="-")
    
file.columns=['nombre', 'correo', 'fecha','edad', 'estatura', 'peso', 'estrato', 'h_estudio', 'pasear', 'deporte', 'caminar', 'tv', 'cine', 'pasear', 'deporte', 'caminar', 'tv', 'cine']
```
